```python
plt.figure(figsize=(8,6))
ax = sns.scatterplot(data=df2, 
                     x="revenue", y="visitors", hue="category", palette="Reds")
sns.move_legend(
    ax, "lower center",
    bbox_to_anchor=(0.5, -.4),
    ncol=3,
    title=None, frameon=False,
)
plt.xlabel("revenue 2020")
plt.ylabel("monthly visitors")
plt.savefig("Seaborn_move_legend_to_the_bottom_of_plot.png",
                    format='png',dpi=150)
```
